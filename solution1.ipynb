# json file to Readable format
 

```python
import json
```

```python
def load_data(filename):
    with open(filename, 'r') as f:
        data = json.load(f)
    return data
```

```python
data=load_data('data1.json')
for user in data["users"]:
    print(user)
    print("\n")
```

```python
def display_connections(data):
    print("Connections:\n")
    for user in data["users"]:
        print(f"ID {user['id']}:- {user['name']} is friend with {user['friends']} and liked pages are {user['liked_pages']}\n")
    print("\nPages:\n")    
    for page in data["pages"]:
       print(f"ID {page['id']}:- {page['name']} is liked by {page['name']}\n")
display_connections(data)
```
